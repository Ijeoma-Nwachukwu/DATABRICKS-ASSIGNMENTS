```
import re

def convert_name(filepath):
    m= re.match("^.*([0-9]{2})-([0-9]{2})-([0-9]{4}).csv$",filepath)
    return int(m.group(3)+m.group(1)+m.group(2))
```

```
csvPairRDD=csvFilesRDD.map(lambda t: (convert_name (t),t))
```

```
csvPairRDD.take(5)
```

```
mostrecentRDD =csvPairRDD.sortByKey(ascending=False)
mostrecentRDD.take(5)
```
