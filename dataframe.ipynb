```
from pyspark.sql.types import *

mySchema = StructType ([StructField ("index", IntegerType()), StructField("webpage", StringType()), StructField("associated_files", StringType())])

myRDD = sc.textFile("/FileStore/webpage/*").map(lambda line: line.split("\t")).map(lambda values: [int(values[0]), values[1], values[2]])
                     
                     
webpages = spark.createDataFrame(myRDD, mySchema)

webpages.show(10)
```
